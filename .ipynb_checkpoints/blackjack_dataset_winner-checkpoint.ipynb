```python
import pandas as pd
import numpy as np
df = pd.read_csv('blkjckhands.csv', sep=',')
df
```

```python
#find the data which does not follow the game rule
invalid_data = df.loc[(df['card3'] == 0) & ((df['card4'] != 0) | (df['card5'] != 0))]
invalid_data_index = df.loc[(df['card3'] == 0) & ((df['card4'] != 0) | (df['card5'] != 0))].index
invalid_data_index
```

```python
#remove invalid data
df = df.drop(invalid_data_index)
```

```python
df=df.replace({'winloss': {'Win': 1, 'Loss': -1, 'Push':0}})
```

```python
df
```

```python
df.groupby(['PlayerNo']).sum()
```

```python
df.groupby(['PlayerNo'])['winloss'].count()
```

```python
win_rate = df.groupby(['PlayerNo'])['winloss'].sum()/df.groupby(['PlayerNo'])['winloss'].count()
win_rate
```

```python
#find the win most player
win_rate.idxmax()
```
